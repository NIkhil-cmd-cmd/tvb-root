````python
import scipy.stats as stats
````

````python
def plot_normal(mu, sigma, ax, **kwds):
    x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
    ax.plot( x, stats.norm.pdf(x, mu, sigma), **kwds )
````
